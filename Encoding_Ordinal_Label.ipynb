Encoding Categorical Data - Ordinal

Ordinal Encoding:
A technique to convert categorical text data into numbers, but in a specific, meaningful order.
Education Level: ['High School', 'Bachelor', 'Master', 'PhD']



Label Encoding: Used mainly for target variable (y), not features.

Ordinal Encoding

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/customer.csv')
```

```python
df.head()
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2],
                                                    df.iloc[:,-1],  # Target (last column)
                                                    test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School','UG', 'PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
oe.categories_
```

Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
