Column Transformer

```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('/content/covid_toy.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (df.drop(columns=['has_covid']), df['has_covid'], test_size=0.2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
```

```python
# Creating a column transformer to preprocess multiple columns in one step
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(), ['fever'] ),
     ('tnf2', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough']) ,
     ('tnf3', OneHotEncoder(sparse_output=False, drop='first'), ['gender', 'city'] )

], remainder= 'passthrough')

# Transformer 1: Apply SimpleImputer to the 'fever' column to fill missing values
# Transformer 2: Apply OrdinalEncoder to the 'cough' column
# Transformer 3: Apply OneHotEncoder to 'gender' and 'city' columns

    # drop='first' avoids the dummy variable trap (drops first category of each)
    # sparse_output=False returns a dense NumPy array instead of a sparse matrix

# Leave any other columns (like 'age') unchanged and pass them through
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
